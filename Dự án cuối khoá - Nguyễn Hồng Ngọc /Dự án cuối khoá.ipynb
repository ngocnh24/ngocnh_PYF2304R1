```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox

class StudentRegistrationSystem:
    def __init__(self, app):
        self.app = app
        self.app.title("Student Registration System")
        self.std_list = [["001", "Ngoc", "Nguyen", "037", "python, C"], ["002", "Anh", "Tran", "098", "python, C"], ["003", "Lan", "Hoang", "016", "python, C"], ["004", "Thang", "Dinh", "078", "python, C"]]
        
#Tạo phần Student Information

        self.detail_frame = ttk.LabelFrame(app, text="Student Information")
        self.detail_frame.grid(row=0, column=0)
        
        self.lbl_id = ttk.Label(self.detail_frame, text="Student ID: ")
        self.lbl_id.grid(row=0, column=0)
        self.entry_id = ttk.Entry(self.detail_frame, width=30)
        self.entry_id.grid(row=0, column=1)
        
        self.lbl_first = ttk.Label(self.detail_frame, text="First Name: ")
        self.lbl_first.grid(row=1, column=0)
        self.entry_first = ttk.Entry(self.detail_frame, width=30)
        self.entry_first.grid(row=1, column=1)
        
        self.lbl_last = ttk.Label(self.detail_frame, text="Last Name: ")
        self.lbl_last.grid(row=2, column=0)
        self.entry_last = ttk.Entry(self.detail_frame, width=30)
        self.entry_last.grid(row=2, column=1)
        
        self.lbl_phone = ttk.Label(self.detail_frame, text="Phone: ")
        self.lbl_phone.grid(row=3, column=0)
        self.entry_phone = ttk.Entry(self.detail_frame, width=30)
        self.entry_phone.grid(row=3, column=1)
        
        self.lbl_course = ttk.Label(self.detail_frame, text="Courses: ")
        self.lbl_course.grid(row=4, column=0)
        self.entry_course = ttk.Entry(self.detail_frame, width=30)
        self.entry_course.grid(row=4, column=1)
        
#Tạo các nút lệnh 

        self.button_frame = ttk.Frame(app)
        self.button_frame.grid(row=5, column=0)
        
        self.btn_register = ttk.Button(self.button_frame, text="Register", command=self.register_std)
        self.btn_register.grid(row=0, column=0)
        
        self.btn_update = ttk.Button(self.button_frame, text="Update", command=self.update_data)
        self.btn_update.grid(row=0, column=1)
        
        self.btn_delete = ttk.Button(self.button_frame, text="Delete", command=self.del_data)
        self.btn_delete.grid(row=0, column=2)
        
        self.btn_clear = ttk.Button(self.button_frame, text="Clear", command=self.clear_data)
        self.btn_clear.grid(row=0, column=3)
        
# Tạo phần tìm kiếm bằng ID
        self.searchid_frame = ttk.LabelFrame(app, text="Search by ID")
        self.searchid_frame.grid(row=7, column=0)
        
        self.lbl_search_id = ttk.Label(self.searchid_frame, text="Type ID here:")
        self.lbl_search_id.grid(row=0, column=0)
        self.entry_search = ttk.Entry(self.searchid_frame)
        self.entry_search.grid(row=0, column=1)
        
        self.btn_search_id = ttk.Button(self.searchid_frame, text="Search", command=self.search_id)
        self.btn_search_id.grid(row=0, column=2)
        
# Tạo phần table thông tin đăng ký 
        self.record_frame = ttk.LabelFrame(app, text="Registration Record")
        self.record_frame.grid(row=9, column=0)
        
        self.record_table = ttk.Treeview(self.record_frame)
        self.record_table.pack(fill=tk.X, pady=5)
        
        self.record_table.bind("<<TreeviewSelect>>", self.entry_student)
        
        self.record_table['columns'] = ['ID', 'First Name', 'Last Name', 'Phone', 'Course']
        self.record_table.column("#0", anchor=tk.W, width=0, stretch=tk.NO)
        self.record_table.column("ID", anchor=tk.W, width=50)
        self.record_table.column("First Name", anchor=tk.W, width=100)
        self.record_table.column("Last Name", anchor=tk.W, width=100)
        self.record_table.column("Phone", anchor=tk.W, width=100)
        self.record_table.column("Course", anchor=tk.W, width=100)
        
        self.record_table.heading("ID", text="ID", anchor=tk.W)
        self.record_table.heading("First Name", text="First Name", anchor=tk.W)
        self.record_table.heading("Last Name", text="Last Name", anchor=tk.W)
        self.record_table.heading("Phone", text="Phone", anchor=tk.W)
        self.record_table.heading("Course", text="Course", anchor=tk.W)
        
        self.load_std_data()
        
# Các chức năng
    
    def load_std_data(self):
        for item in self.record_table.get_children():
            self.record_table.delete(item)
        
        for i in range(len(self.std_list)):
            self.record_table.insert(parent="", index="end", text="", iid=i, values=self.std_list[i])
    
    def entry_student(self, event):
        selected_item = self.record_table.selection()[0]
        student = self.std_list[int(selected_item)]
        
        self.entry_id.delete(0, tk.END)
        self.entry_first.delete(0, tk.END)
        self.entry_last.delete(0, tk.END)
        self.entry_phone.delete(0, tk.END)
        self.entry_course.delete(0, tk.END)
        
        self.entry_id.insert(0, student[0])
        self.entry_first.insert(0, student[1])
        self.entry_last.insert(0, student[2])
        self.entry_phone.insert(0, student[3])
        self.entry_course.insert(0, student[4])
    
    def register_std(self):
        std_id = self.entry_id.get()
        std_first = self.entry_first.get()
        std_last = self.entry_last.get()
        std_phone = self.entry_phone.get()
        std_courses = self.entry_course.get()
        
        
        self.std_list.append([std_id, std_first, std_last, std_phone, std_courses])
        self.load_std_data()
    
    def update_data(self):
        selected_item = self.record_table.selection()[0]
        i = int(selected_item)
        
        std_id = self.entry_id.get()
        std_first = self.entry_first.get()
        std_last = self.entry_last.get()
        std_phone = self.entry_phone.get()
        std_courses = self.entry_course.get()
        
        self.std_list[i] = [std_id, std_first, std_last, std_phone, std_courses]
        self.load_std_data()
    
    def del_data(self):
        selected_item = self.record_table.selection()[0]
        i = int(selected_item)
        
        del self.std_list[i]
        self.load_std_data()
    
    def clear_data(self):
        self.entry_id.delete(0, tk.END)
        self.entry_first.delete(0, tk.END)
        self.entry_last.delete(0, tk.END)
        self.entry_phone.delete(0, tk.END)
        self.entry_course.delete(0, tk.END)
        self.load_std_data()
    
    def search_id(self):
        search_id = self.entry_search.get()
        found = False
        
        for i in range(len(self.std_list)):
            if self.std_list[i][0] == search_id:
                self.record_table.selection_set(i)
                self.record_table.focus(i)
                self.record_table.see(i)
                found = True
                break
        
        if not found:
            messagebox.showinfo("Search Result", "No matching student ID found.")
    
    def run(self):
        self.app.mainloop()


if __name__ == "__main__":
    app = tk.Tk()
    system = StudentRegistrationSystem(app)
    system.run()

  
```

```python
class Book: 
    def __init__(self, title, genre, price):
        self.title = title 
        self.genre = genre 
        self.price = price 
    def __str__(self):
        return f"Title: {self.title}, genre: {self.genre}, price: {self.price}" 
    
```

```python
from tkinter import* 
from tkinter import messagebox

window = Tk()
window.title("Bookstore Management")

book_list = []

def add_book(): 
    book_title = title.get()
    book_genre = genre.get()
    book_price = price.get()
    
    if book_title == "" or book_genre =="" or book_price =="": 
        messagebox.showerror("Book", "Please enter all details")
    else: 
        new_book = Book(book_title, book_genre, float(book_price))
        book_list.append(new_book)
    print([book.title for book in book_list])
    messagebox.showinfo("Book", f"{book_title} added successfully!")

def show_info_book(): 
    search_book = search_name.get()
    check = False 
    
    for i in range(len(book_list)): 
        book_info = Book(book_list[i].title, book_list[i].genre, book_list[i].price)
        if search_book == book_list[i].title:
            messagebox.showinfo("Book Search", book_info )
            check = True 
            break 
    if check == False:
        messagebox.showerror("Book Search", "Book is not in the list!")
    
def delete_book():
    delete_book = search_name.get()
    check = False 
    
    for i in range(len(book_list)): 
        if delete_book == book_list[i].title: 
            messagebox.showinfo("Book Delete", f"{delete_book} deleted successfully!")
            check = True 
            break 
        if check == False: 
            messagebox.showerror("Book Delete", "Book is not in the list!")

def update_price():
        book = search_name.get()
        price = update_price.get()
        check = False 
        
        for i in range(len(book_list)): 
            if book == book_list[i].title: 
                float(price) = book_list[i].price
                messagebox.showinfo("Price Update", f"Price of {book} updated successfully!")
                check = True 
                break 
            else: 
                check == False: 
                messagebox.showerror("Price Update", "Book is not in the list!")
            

lbl_name = Label(window, text = "Title: ")
lbl_genre = Label(window, text = "Genre: ")
lbl_price = Label(window, text = "Price: ")
lbl_search = Label(window, text = "Book: ")
lbl_update = Label(window, text = "Update Price: ")

title = Entry(window)
genre = Entry(window)
price = Entry(window)
search_name = Entry(window)
update_price = Entry(window)  

btn_add = Button(window, text = "Add Book", command = add_book)
btn_search = Button(window, text = "Search", command = show_info_book)
btn_del = Button(window, text = "Delete", command = delete_book)
btn_update = Button(window, text = "Update Price", command = update_price)

lbl_name.grid(row = 0, column = 0)
title.grid(row = 0, column = 1)
lbl_genre.grid(row = 1, column = 0)
genre.grid(row = 1, column = 1)
lbl_price.grid(row = 2, column= 0)  
price.grid(row = 2, column= 1)  
btn_add.grid(row = 3, column= 1)              
lbl_search.grid(row = 4, column =0)
search_name.grid(row = 4, column =1)
btn_search.grid(row = 4, column =2)              
btn_del.grid(row = 4, column =3)
lbl_update.grid(row = 5, column =0)
update_price.grid(row = 5, column =1)
btn_update.grid(row = 5, column =2) 

window.mainloop()              
```
